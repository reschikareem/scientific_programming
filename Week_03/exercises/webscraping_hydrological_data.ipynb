# Web Scraping Hydrological Data with Python and Beautifulsoup

Data from: https://hydroproweb.zh.ch/Listen/AktuelleWerte/aktuelle_werte.html

## Libraries and settings

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

## Web Scraper Hydrological Data

```python
# Send an HTTP request to the URL
url = 'https://hydroproweb.zh.ch/Listen/AktuelleWerte/aktuelle_werte.html'
response = requests.get(url)
html_content = response.content

# Parse the HTML content
soup = BeautifulSoup(html_content, 'html.parser')

# Locate the table and extract values
table = soup.find('table')

# Extract table headers
headers = [header.get_text().replace('\xa0', '') for header in table.find_all('th')]

# Extract table rows
rows = []
for row in table.find_all('tr')[1:]:
    cells = row.find_all('td')
    row_data = [cell.get_text().replace('\xa0', '').strip() for cell in cells]
    rows.append(row_data)

# Split the combined "time_date" column into separate "Zeit" and "Datum" columns
clean_rows = []
for row in rows:
    if len(row) == 9:  # Ensure the row has the correct number of columns
        time_date = row[2]
        time, date = time_date[:5], time_date[5:]
        clean_row = row[:2] + [time, date] + row[3:]
        clean_rows.append(clean_row)

# Define the final columns
columns = ['water_body',
            'unit',
            'time',
            'date',
            'current_value',
            'value_24h_ago',
            'difference',
            'mean_24h',
            'max_24h',
            'min_24h']

# Create DataFrame
df = pd.DataFrame(clean_rows, columns=columns)

# Print the DataFrame
df
```
